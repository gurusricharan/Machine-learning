```python
X = sm.iloc[:,[11,29]].values
```

```python
from sklearn.cluster import KMeans
```

```python
elbow = []
for i in range(1,15):
    cf = KMeans(n_clusters = i)
    cf.fit(X)
    elbow.append(cf.inertia_)
    
```

```python
elbow
```

```python
plt.plot(range(1,15),elbow)
```

```python
guru = KMeans(n_clusters = 5)
guru.fit(X)
```

```python
pred = guru.predict(X)
```

```python
pred
```
